```python
import numpy as np

def generateX(dataSize):
    x=np.random.uniform(1,-1,dataSize)
    return x

def sign(x):
    if x<=0:
        return -1
    if x>0:
        return 1

def generateY(x,dataSize,noise_rate):
    #get s(x)
    i=0
    y=np.zeros(dataSize)
    for i in range(dataSize):
        y[i]=sign(x[i])
    
    #get noise
    j=0
    randomList=[]
    randomList=np.random.rand(dataSize)
    for j in range(dataSize):
        if (randomList[j]/np.max(randomList))<noise_rate:
            y[j]=-y[j]
    
    return y

def generateTheta(x):
    Xlen=len(x)
    xList=sorted(x)
    theta=np.zeros(Xlen)
    
    for i in range(Xlen-1):
        theta[i]=(xList[i]+xList[i+1])/2
        
    theta[-1]=1
    
    return theta

def caculateError(x,y,theta,s):
    nx=len(x)
    error=0
    for i in range(nx):
        temp=s*sign(x[i]-theta)
        if temp!=y[i]:
            error=error+1
    return error

def deciStump(X,Y,exTimes):
    E_in=0
    E_out=0
    for j in range(exTimes):
        print(j)
        dataSize=len(X)
        theta=generateTheta(X)
        e_in=np.zeros((2,dataSize))
        for i in range(len(theta)):
            e_in[0][i]=caculateError(X,Y,theta[i],1)/dataSize
            e_in[1][i]=caculateError(X,Y,theta[i],-1)/dataSize
            s=0
            theta_best=0
        
        mins1, mins01=np.min(e_in[0]), np.min(e_in[1])
        if mins1<mins01:
            s=1
            theta_best=theta[np.argmin(e_in[0])]
        else:
            s=-1
            theta_best=theta[np.argmin(e_in[1])]
            
        e_out=0.5+0.3*s*(np.abs(theta_best)-1)
        E_in=E_in+np.min(e_in)
        E_out=E_out+np.min(e_out)
    
    avg_Ein=E_in/exTimes
    avg_Eout=E_out/exTimes
    return avg_Ein, avg_Eout

def getData(file_root):
    b=np.array(np.loadtxt('hw2_train.dat.txt'))
    dataShape=b.shape
    xLen=b.shape[1]-1
    xNum=b.shape[0]
    X=b[:,0:xLen]
    Y=b[:,xLen:xLen+1]
    
    return X,Y

def getXi(X,i):
    Xi=X[:,i]
    Xi=np.transpose(Xi)
    
    return Xi

file_root_train="hw2_train.dat.txt"
file_root_test="hw2_test.dat.txt"

exTimes=100
noiseRate=0.2


X_train,Y_train=getData(file_root_train)

E_in1=np.zeros((1,X_train.shape[1]))
E_out1=np.zeros((1,X_train.shape[1]))
#print(X_train.shape[1])
for i in range(X_train.shape[1]):
    Xi=getXi(X_train,i)
    E_in1[0,i],E_out1[0,i]=deciStump(Xi,Y_train,exTimes)

print(np.min(E_in1),np.min(E_out1))
```
