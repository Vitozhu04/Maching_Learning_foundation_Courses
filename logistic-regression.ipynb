```python
import numpy as np
import math

def sign(x1,x2):
    if x1**2+x2**2-0.6<=0:
        return -1
    else:
        return 1

def calError(feature,w,y):
    errorCount=0
    for i in range(len(y)):
        if np.dot(feature[i],w)*y[i]<0:
            errorCount += 1
    
    return errorCount/len(y)

def getData(file_root):
    b=np.array(np.loadtxt(file_root))
    dataShape=b.shape
    xLen=b.shape[1]-1
    xNum=b.shape[0]
    X=b[:,0:xLen]
    #
    Xnew=np.ones((dataShape))
    Xnew[:,1:len(Xnew[0])]=X
    Y=b[:,xLen:xLen+1]
    
    return Xnew,Y

def getXi(X,i):
    Xi=X[:,i]
    Xi=np.transpose(Xi)
    
    return Xi

def fixedGD(x,y,w):
    temp=-y*(np.dot(x,w))
    theta=np.exp(temp)/(1+np.exp(temp))
    gradient=(1/len(x))*(sum(theta*-y*x).reshape(len(w), 1))
    return gradient

def sgd(x,y,w):
    temp=-y*(np.dot(x,w))
    theta=np.exp(temp)/(1+np.exp(temp))
    gradient=np.mat(theta*y*x)
    print(gradient)
    gradient=gradient.reshape(len(w), 1)
    print(gradient)
    return gradient

```
